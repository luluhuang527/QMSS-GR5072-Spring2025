## Question 3
### [10 pts] Which player had the best free throw percentage ("FT%") from the year 2000 to the most recent year in the data? If there are multiple players who have tied for the best percentage, then sort on the column with player names and submit your answer as a pd.DataFrame object.

```python
# 3. Best free throw percentage from 2000 onward
# Filter the data for the year 2000 onward and sort by FT% (descending) and Player (ascending)
best_ft = df[df['Year'] >= 2000].sort_values(by=['FT%', 'Player'], ascending=[False, True])

# Find the highest FT% in the sorted data
max_ft = best_ft['FT%'].max()

# Filter the players with the highest FT% and return the result as a DataFrame
best_ft_players = best_ft[best_ft['FT%'] == max_ft]

#Display relevant columns of first and last 5 rows
display(best_ft_players[['Player', 'FT%', 'Year']].head(), best_ft_players[['Player', 'FT%', 'Year']].tail())
```

```python
best_ft_players.info()
```

### From the year 2000 onwards, 313 players tied for the free throw percentage of 100%. They are listed in alphabetical order in the data frame best_ft_players.

## Question 4
### [10 pts] Among those with at least 50 attempts ('FTA'), which player had the best free throw percentage ("FT%") from the year 2000 to the most recent year in the data?

```python
# 4. Best free throw percentage with at least 50 attempts from 2000 onward
best_ft_50 = df[(df['Year'] >= 2000) & (df['FTA'] >= 50)].sort_values(by=['FT%','Player'], ascending=[False, True])

# Find the highest FT% in the sorted data
max_ft_50 = best_ft_50['FT%'].max()

# Filter the players with the highest FT% and return the result as a DataFrame
best_ft_50_players = best_ft_50[best_ft_50['FT%'] == max_ft_50]

#Display relevant columns of first and last 5 rows
display(best_ft_50_players[['Player', 'FT%', 'FTA', 'Year']].head(1))
```

### Among those with at least 50 attempts from the year 2000 onwards, Terrell Brandon had the best free throw percentage of 98.8%.

## Question 10
### [10 pts] Subset the original dataset to the years 1995-1997. Group the data by team name and year and then summarize the average number of points per team. Arrange from most to least points. (Hint: Your answer should have a shape of 88 x 3, since we are looking at the average points scored over the 1995-1997 seasons. Note that 2 of these columns could be indexes, in which case Pandas will tell you your data is 88 rows x 1 columns).

```python
# 10. Subset data for 1995-1997 and summarize average points per team
# Subset data for 1995-1997
df_95_97 = df[(df['Year'] >= 1995) & (df['Year'] <= 1997)]

# Group by team name and year, and calculate average points per team
team_average_pts = df_95_97.groupby(['Tm', 'Year'])['PTS'].mean().reset_index()

# Sort the data by average points in descending order
team_average_pts.sort_values(by='PTS', ascending=False, inplace=True)

# Show the result
team_average_pts
```

## Question 11
### [10 pts] Similar to Q10, start by subsetting the original dataset to the years 1995-1997. Then create a wide dataset that keeps each year in its own row, but spreads team names to multiple columns with each column delineating points per team in each of the 1995-1997 seasons (Hint: Your final answer should have a shape of 3 x 30).

```python
# 11. Create wide format dataset for 1995-1997 points per team
df_wide = team_average_pts.pivot(index='Year', columns='Tm', values='PTS')
```

## Question 12
### [10 pts] Now return the data from #11 back into a long (tidy) format by moving Team (TM), Points and Year into their own columns (Hint: Your data shape should be 90 x 3).

```python
# 12. Convert wide dataset back to long format
df_long = df_wide.reset_index().melt(id_vars=['Year'], var_name='Team', value_name='PTS')
```
